```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
ARTICLE = """
Artificial Intelligence (AI) is the simulation of human intelligence in machines designed to think, learn, and solve problems. Leveraging algorithms and vast amounts of data, AI can recognize patterns, make decisions, and adapt to new information, often surpassing human efficiency in specific tasks. It underpins technologies such as natural language processing, computer vision, and robotics, enabling applications like virtual assistants, personalized recommendations, and autonomous vehicles. AI continues to transform industries, from healthcare and finance to education and entertainment, driving innovation while raising ethical considerations about privacy, bias, and the future of work.







"""
```

```python
summarizer(ARTICLE, max_length=130, min_length=30, do_sample=False)
```
